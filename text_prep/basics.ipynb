# REMOVING HTML tags

```python
e = """<!DOCTYPE html> <html>
<head>
<style>
  .custom-hr {
    border: none; /* Remove default border */
    border-top: 2px dashed red; /* Set top border for the dashed line */
    width: 50%; /* Set width to 50% of the parent element */
    margin: 20px auto; /* Center the line and add vertical margin */
  }
</style>
</head>
<body>

<p>Some text above the custom line.</p>
<hr class="custom-hr">
<p>Some text below the custom line.</p>

</body>
</html>"""
```

```python
z = re.compile('<.*?>')
print(z.sub(r'', e))
```

#  Tokenization using spacy

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")  # it is pre trained model for various nlp tasks
doc = nlp("hello what's the hell is going is on")
for i in doc:
    print(i, " | ",
          spacy.explain(i.pos_),
          " | ", i.lemma_)
```
